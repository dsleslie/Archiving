```python
df=pd.read_csv('c:/temp100/house_price.csv')
df
```

```python
df.columns
```

```python
df.drop('Unnamed: 0', axis=1)
# df: unnamed:0번 을 가지고 있습니다 


df1=df.drop('Unnamed: 0', axis=1)
df1  # 삭제된 버전이 됩니다. 
```

```python
# 분석에서 사용한다. : df1 
```

```python
# 데이터 프레임의 일부분을 가져오고자 한다. 
# df1.iloc, df1.loc  이런 함수를 이용해서 주어진 데이터의 일부분을 가져올 수 있다. 
df1.iloc[2:4] 
```

```python
df1.iloc[2:4, 1:3]  # 앞에 번호는 row, 뒤에 번호는 column 번호 
```

```python
df1.loc[df1.index[2:5],['month','price']]  # loc는 row와 column의 label을 이용해서 가져온다. df.index 
```

```python
df1.info()
```

```python
df1['size_p']=df1['size']/3.3  # 아파트 면적으로 평으로 계산한 것이다. 
df1['price_p']=df1['price']/df1['size_p']
df1
```

```python
df1.describe() ## data frame에서 describe()
```

```python
df1['price'].mean()  # 숫자변수이므로 평균이 계산된다. 
```

```python
df1.mean() # 문자 변수가 포함되어 있기 때문에 
```

```python
# plt의 hist 함수를 사용한다. 
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rcParams
from matplotlib import font_manager, rc 



```

```python
plt.hist(df['price_p'], bins=50, color='g')
# x축의 레이블을 설정하고 싶다. "평당 가격"
plt.xlabel("평당가격")
plt.grid()
plt.show()
```

```python
fig, axs=plt.subplots(1,2) # 2개 그릴 것이다. 
fig.set_size_inches(9,3)  # 가로 세로 길이를 결정한다. 

# 코드 수정 예정 : 아래와 같이 "="가 아니라 .(period)를 찍어야 합니다. 
axs[0].hist(df['price_p'], bins=30, density=True, color='red', cumulative=True) # 누적확률을 계산한다. 
axs[1].hist(df['price_p'], bins=30, density=True, color='blue')
plt.show()
```
