```python
from urllib.request import urlopen   # request package 
from bs4 import BeautifulSoup  # bs4 package
import pandas as pd
```

```python
ServiceKey="29jTU1S%2B7bDI1Rb2LV0S5D0d3Xn%2Fji9T3k%2FWjsLR%2FyI9O3wEU0atKsr9hj9ATsfIpKgWmV8k8Ll67O%2F73Nyj6w%3D%3D"
month=202401 # 2024년 1월 거래 정보를 사용하고자 한다. 
gu_code=11110 # 서울시 종로구 데이터를 사용하겠다. 
numOfRows=1000  # 최대 1000개 거래정보를 가져오겠다. 
url="http://openapi.molit.go.kr:8081/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTrade?LAWD_CD="+str(gu_code)+"&DEAL_YMD="+str(month)+"&numOfRows="+str(numOfRows)+"&serviceKey="+ServiceKey
```

```python
print(url)
```

```python
result=urlopen(url)
house=BeautifulSoup(result, 'lxml-xml')
house
```

```python
te=house.find_all("item")
te
```

```python
data=[]
for i in range(len(te)):
    price=te[i].거래금액.string.strip()
    apt_name=te[i].아파트.string.strip()
    dong_name=te[i].법정동.string.strip()
    
    total=[price,apt_name, dong_name]
    data.append(total)

print(data)  #  list type의 데이터로 저장이 된다. 
    
```

```python
# 마지막으로 저장된 list "data"를 excel 또는 csv 파일로 내보낸다. 
# csv 파일로 저장해보자. 
# 1단계: list를 데이터 프레임으로 생성한다. pandas 의 data frame으로 변환한다. 
# 2단계: pandas data frame을 csv파일로 내보낸다. 
# data frame 이름을 temp로 지정한다. 
temp=pd.DataFrame(data, columns=['price', 'apt_name', 'dong_name']) # columns는 데이터 프레임에서 각 column의 이름을 지정한다. 
temp 
```

```python
temp.to_csv("c:/temp100/house_price100.csv")  # csv 파일을 열어보면 한글이 깨져보인다. 
# 그러나 해당 파일을 다시 python으로 불러오면 한글이 정상으로 나온다. 
```

```python
temp.to_csv("c:/temp100/house_price100.csv",encoding='utf-8-sig')  # csv 파일에서 한글이 깨져 보이는 것이 싫으신 분들은 encoding 옵션을 사용한다.
```
