```python
# random drawing with replacement  : 복원 추출
# random drawing without replacement : 비복원추출
# random package 
```

```python
import random 
```

```python
box=[1,2,3,4,5] # list object
box
```

```python
random.seed(123)
sample_with_replacement=random.choices(box, k=2) # with replacement 명령어 
sample_with_replacement
```

```python
# without replacement : 비복원추출
random.seed(42)
sample_without_replacement=random.sample(box, k=2) 
sample_without_replacement
```

```python
for i in range(10):
    sample_without_replacement=random.sample(box, k=2) 
    print(sample_without_replacement)
    
```
