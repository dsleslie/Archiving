### 상관계수 correlation coefficient
- 두 변수가 같은 방향으로 움직이는 지>0, 다른 방향으로 움직이는 지<0
- Pearson's correlation
- 상관계수 계산법
- correlation(상관관계) 과 causality(인과관계)의 차이
    - 상관관계엔 원인과 결과가 없음. 관계가 있다. 그냥
    - 인과관계는 원인과 결과가 있는 관계
    - 상관관계가 더 넓은 개념

```python
# 두 변수의 상관계수 구해보자
data2[['chnimp','bchlimp']].corr()
```

```python
# 0.27206 -> 양수, 두 변수 정비례관계, 양의 상관관계를 가지고 있다.
```

### numpy 모듈의 활용

- Series라는 함수를 사용하면 직접 데이터를 입력할 수도 있다

```python
cashflow=pd.Series(index=[0,1,2,3,4,5],name='Cashflow',data=[-107,5,5,5,5,105])
cashflow
# 정부가 107억 투자, 5억씩 번거 .. 초반엔 수익이 잘 안난다. 마지막 해에 105억 나는 '투자안'
# 투자해야할까?
# 107억 투자해서 125억 버는데, 할인 적용해서 '순현재가치 npv' 를 보고 판단해야함
```

### 내부수익률 IRR Internal Rate of Return 계산하는 방법

- 투자안을 보고 투자를 결정하는 방법 중 하나
- 내부수익률 정의: npv를 0으로 만드는 할인율 = IRR
- 내부수익률이 높을 수록 투자해야함.. 
- 이자비용보다 IRR이 높다면 '투자 선택 !'
- numpy 패키지 안에 안에 IRR 계산해주는 함수 있음
- numpy, numpy_financial as npf

```python
# IRR 계산하는 함수 : npf.irr()

npf.irr(cashflow)
```
