### Bagging

- 전체 iris 150개 데이터를 train test set으로 나누자
- 먼저 데이터 70%을 train으로 하고, 나머지 30% test set 으로 둔다
- train set = 150*0.7 = 105
- bagging을 105개 무작위로 여러번 뽑는다. 뽑을 때마다 BS1->tree1


```python
X=iris[['Sepal.Length','Sepal.Width']]
y=iris['Species']
```

```python
X
```

```python
y
```

```python
# train set과 test set으로 구분하자

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=0)
X_train
```

```python
X_test
```

```python
# train set으로 Bagging한다.

# y 가 범주형 -> RandomForestClassifier

bag1=RandomForestClassifier(n_estimators=10,max_features=None,random_state=0,max_depth=2)
```

```python
bag1
```

```python
result1=bag1.fit(X_train,y_train)
result1
```

```python
pred1=result1.predict(X_test)
# 위에 n_estimators 가 10이니까 지금 bagging으로 10개 나무 만들어진 거고, 
# 그걸 test set 넣어서 얼만큼 잘 예측하는지 보려는 것
```

```python
pred1
```

```python
confusion_matrix(y_test,pred1)
```
