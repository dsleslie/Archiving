```python
# 위에 describe 본 벼눗는 다 연속형 변수
# diagnosis 와 id를 제외한 위의 30개 변수는 다 x 변수. 거리를 측정하기 위해 사용하는 것들.
# 우선 x 변수를 지정해야한다. 
# x 변수가 몇개 안되면 위에 describe 에 썼던 것 처럼 [[]] 해서 쓰면되는데, 지금은 30개를 다 쓰기 힘듦.

# 난 2번째 부터 32번째 까지 다  x 변수 하고 싶다. 어디서부터 어디까지 라고 나타낼 수 없나?
# 첫번째 : 전체부터 다 가져와라 
# 왜 32번째? : 
# [:,2:32] : (0,1,)2번부터 앞에서 1부터 셌을 때 32번째 위치까지 가져와라 



X1=bc_data.iloc[:,2:32]
X1
```

```python
# min max normalize

scaler=preprocessing.MinMaxScaler()
#minmaxnormalize 된 변수 X11
X11=scaler.fit_transform(X1)
X11
# 다 0과 1사이로 표준화됨.
```

```python
# 정규화
X12=preprocessing.scale(X1)
X12
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
y=bc_data['diagnosis']
X_train, X_test, y_train, y_test=train_test_split(X11,y,train_size=0.7,random_state=0)
X_train
```

```python
# Train set 이용해서 KNN 실행

# 우선 K를 정하자. 안정하면 그냥 default를 1로 함.
# K를 3으로 하자 : 가까운 3명 통해 판단

clf=KNeighborsClassifier(n_neighbors=3)
result=clf.fit(X_train,y_train)
result
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
len(y_pred)
# test의 171개에 대한 예측값을 제시한 것.
```

```python
# y_test와 y_pred 간에 얼마나 매칭 되는지 알아보자 by confusion matrix의 ccr

# 이 패키지를 사용해서 구한 것: from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
# 올바르게 분류한 정도가 97%이다
(111+55)/171
```
