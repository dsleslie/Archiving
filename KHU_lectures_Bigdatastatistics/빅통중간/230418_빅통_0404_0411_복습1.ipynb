# Logistic regression



```python
# logistic regression 의 cdf 그리기 1
x=np.linspace(-5,5,100)
y=np.exp(x)/(1+np.exp(x))
plt.plot(x,y)
plt.show()
```

```python
# logistic regression 의 cdf 그리기 1
x=np.linspace(-5,5,100)
y=logistic.cdf(x)
plt.plot(x,y)
plt.show()
```

```python
# 승산 odds

file1="c:/temp/credit_LN4.csv"
credit=pd.read_csv(file1)

y=credit['default']
x=credit['balance']
X=sm.add_constant(x)

y=pd.get_dummies(y)
y=y['Yes']

smlogit=sm.Logit(y,X).fit()
smlogit.summary()
```

```python
# y가 1일 확률 구하자

# 상수항이 1일 때, balance x값은 1000일 때
X1=[1,1000]

smlogit.predict(X1)
```

```python
X2=[1,2000]
smlogit.predict(X2)
# balance 2배 되니 default할 확률 매우 커짐
```
